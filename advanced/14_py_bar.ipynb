```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


#loading data

dataset=load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()

# Data Cleanup

df['job_posted_date']= pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda skill_list:ast.literal_eval(skill_list)if pd.notna(skill_list) else skill_list)
```

```python
df.head()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=180,autopct='%1.2f%%')
plt.title("Job Titles in Luke's Dataset ")
plt.ylabel("")
plt.show()
```

```python
df[['job_work_from_home','job_no_degree_mention','job_health_insurance']].plot(kind='pie',autopct='%1.2f%%')
```

```python
list_e=['job_work_from_home','job_no_degree_mention','job_health_insurance']
```

```python
fig , ax = plt.subplots(3,1)

for i,item in enumerate(list_e):
    df[item].plot(kind='pie',autopct='%1.2f%%',ax=ax[i])
    
```

```python
fig , ax = plt.subplots(1,3)
dict_column={
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'

}

for i,(column,title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(),autopct='%1.2f%%',labels=['False','True'])
    ax[i].set_title(title)


# ax.pie(df['job_work_from_home'].value_counts())
```
