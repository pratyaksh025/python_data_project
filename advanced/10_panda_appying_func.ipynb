```python
def projected_salary(row):
    if "Senior" in row['job_title_short']:
        return 1.05* row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']

df_salary['salary_year_inflated']=df_salary.apply(projected_salary, axis=1)

df_salary[['job_title_short','salary_year_inflated','salary_year_avg']].sample(5)
```

```python
df_salary['salary_year_inflated_with_lamda']=df_salary.apply(lambda row: 1.05 * row['salary_year_avg'] if "Senior" in row['job_title_short'] else 1.03 * row['salary_year_avg'],axis=1)
```
